### Class Barcelona

```python
class Barcelona:
    def __init__(self, Nheight, Nwidth, pklidx, timestep, xmin=0, xmax=0, ymin=0, ymax=0, df_pos = None, link = None, nbMatrix = None):
        self.Nheight = Nheight
        self.Nwidth = Nwidth
        self.pklidx = pklidx
        self.timestep = timestep
        self.xmin = xmin
        self.xmax = xmax
        self.ymin = ymin
        self.ymax = ymax
        self.df_pos = df_pos
        self.link = link
        self.nbMatrix = nbMatrix

    def gridTensor(self, id):

        df_link = pd.read_csv(r"C:\Users\orane\Documents\EPFL\GC\Master\MA2\Projet LUTS\all_agg\datasets\simbarca\all_agg\metadata\link_bboxes_clustered.csv")

        df_pos = df_link[['id', 'c_x', 'c_y']].copy()
        self.df_pos = df_pos

        # Définition du nombre de lignes et colonnes de la grille
        n_rows, n_cols = self.Nheight, self.Nwidth

        # Déterminer les bornes des points
        xmin1, ymin1 = df_pos[['c_x', 'c_y']].min()
        xmax1, ymax1 = df_pos[['c_x', 'c_y']].max()

        # Ajuster les bordures
        xmin = xmin1 - 0.05*(xmax1-xmin1)
        xmax = xmax1 + 0.05*(xmax1-xmin1)

        ymin = ymin1 - 0.05*(ymax1-ymin1)
        ymax = ymax1 + 0.05*(ymax1-ymin1)

        self.xmin = xmin
        self.xmax = xmax
        self.ymin = ymin
        self.ymax = ymax
        
        # Calculer la taille des cellules de la grille
        cell_width = (xmax - xmin) / n_cols
        cell_height = (ymax - ymin) / n_rows

        # Fonction pour obtenir l'indice de grille d'un point
        def assign_to_grid(x, y):
            col = int((x - xmin) / cell_width)
            row = int((ymax - y) / cell_height)  # Inversion pour que 0 soit en haut
            col = min(col, n_cols - 1)  # S'assurer que les indices restent dans la grille
            row = min(row, n_rows - 1)
            return (row, col)

        # Appliquer l'assignation de grille à chaque point
        df_pos['grid_cell'] = df_pos.apply(lambda row: assign_to_grid(row['c_x'], row['c_y']), axis=1)

        # Afficher le DataFrame résultant dans la console
        print(df_pos)

    def visualizeGrid(self):
        n_cols, n_rows = self.Nheight, self.Nwidth
        # Visualisation
        xmin = self.xmin
        xmax = self.xmax
        ymin = self.ymin
        ymax = self.ymax
        df_pos = self.df_pos
        
        cell_width = (xmax - xmin) / n_cols
        cell_height = (ymax - ymin) / n_rows
        fig, ax = plt.subplots(figsize=(6,6))
        ax.set_xlim(xmin, xmax)
        ax.set_ylim(ymin, ymax)

        # Dessiner la grille
        for i in range(1, n_cols):
            ax.axvline(xmin + i * cell_width, color='gray', linestyle='--')
        for i in range(1, n_rows):
            ax.axhline(ymin + i * cell_height, color='gray', linestyle='--')

        # Afficher les points avec leur ID et leur cellule
        for _, row in df_pos.iterrows():
            ax.scatter(row['c_x'], row['c_y'], color='hotpink')
            #ax.text(row['c_x'], row['c_y'], f"{row['id']}", fontsize=8, verticalalignment='bottom')

        plt.show()

    def getdf_id(self):

        # Dossier où se trouvent les fichiers
        dossier = self.link
        # link = r"C:\Users\orane\Documents\EPFL\GC\Master\MA2\Projet LUTS\all_agg\datasets\simbarca\all_agg"
            
        # Construire le nom du fichier avec le bon format
        nom_fichier = f"agg_timeseries_{self.pklidx}.pkl"
        chemin_fichier = os.path.join(dossier, nom_fichier)
        
        # Vérifier si le fichier existe
        if not os.path.exists(chemin_fichier):
            print(f"❌ Fichier {nom_fichier} non trouvé !")
            return None

        # Charger le fichier pickle
        with open(chemin_fichier, "rb") as f:
            data = pickle.load(f)
        
        return data[id]

    

    def createMatrix(self, timestep, id): # df = df_pos
        df_id = self.getdf_id()
        df = self.df_pos
        n_cols, n_rows = self.Nheight, self.Nwidth

        tensor = np.zeros((n_rows, n_cols))

        for i in range(n_rows):
            for j in range(n_cols):
                somme = 0
                target_cell = (i,j)
                ids_in_cell = df[df['grid_cell'] == target_cell]['id'].tolist()
                for idx in ids_in_cell:
                    somme+= df_id.loc[timestep, idx]
                if id == 'ld_speed':
                    if len(ids_in_cell) == 0:
                        tensor[i,j] = 0 
                    else: 
                        tensor[i,j] = somme/len(ids_in_cell)
                else: 
                    if len(ids_in_cell) == 0:
                        tensor[i,j] = 0 
                    else: 
                        tensor[i,j] = somme
        return tensor
    
    def createTensor3D(self, id):
        df_id = self.getdf_id()
        #print(df_id.shape)
    
        list_timestep = df_id.index.astype(str).tolist()

        maxMatrix = 96 # Nbr max de matrices dans les tenseurs

        listTensor = []
        for i in list_timestep[:maxMatrix]:
            tensor = self.createMatrix(i, id)
            listTensor.append(tensor)

        self.nbMatrix = len(listTensor) # Nbr de matrices "réelles"

        if self.nbMatrix < maxMatrix:
            n_rows, n_cols = self.Nheight, self.Nwidth
            zero_matrix = np.zeros((n_rows, n_cols))  # Matrice remplie de zéros
            listTensor.extend([zero_matrix] * (maxMatrix - self.nbMatrix))  # Ajoût de matrices nulles 

        tensor3D = np.stack(listTensor, axis = 0)

        return tensor3D
    
    def meanTensor(self, tensor3D):

        ogMatrix = tensor3D[:(self.nbMatrix)-1] # Exclut les matrices ajoutées artificiellement
        meanMatrix = np.mean(ogMatrix, axis=(1,2))
        meanValue = np.mean(meanMatrix)

        return meanMatrix, meanValue
```
