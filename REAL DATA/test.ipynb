```python
import os
import numpy as np
import glob

# Chemin vers le dossier contenant les sous-dossiers '000', '001', ..., '099'
root_dir = r"C:\Users\amelo\Documents\Map_configurations\pred_vdist"

# Génère les noms des sous-dossiers
folder_names = [f"{i:03d}" for i in range(100)]

# Liste globale pour stocker tous les tableaux 2D
all_arrays = []

for folder in folder_names:
    folder_path = os.path.join(root_dir, folder)
    # Cherche tous les fichiers t*_map.npy dans le dossier
    npy_files = sorted(glob.glob(os.path.join(folder_path, "t*_map.npy")))
    
    # Tri basé sur l'indice temporel (t0, t1, ..., tN)
    npy_files.sort(key=lambda x: int(os.path.basename(x).split("_")[0][1:]))

    for file_path in npy_files:
        array_2d = np.load(file_path)
        all_arrays.append(array_2d)

# Stack tous les arrays en un seul grand tableau 3D : (N_total, H, W)
result_array = np.stack(all_arrays)

# Sauvegarde si besoin
np.save("stacked_dataset_variable_timesteps.npy", result_array)
```
