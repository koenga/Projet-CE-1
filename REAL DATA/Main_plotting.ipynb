```python
if __name__ == '__main__':

    folder = os.getcwd()

    folder_random = folder + '/Results/06_13_2025_16_29_23 random real presentation'
    folder_greedy = folder + '/Results/06_16_2025_21_43_04_greedy'
    #folder_zz     = folder + '/Results/11_09_2024_18_11_40_zz'
    folder_rl     = folder + '/Results/06_14_2025_01_19_38 GRU real presentation'

    #----- Plot training -----

    # folder_rl1    = folder + '/Results/10_30_2024_00_27_54_pretraining'
    # #folder_rl2    = '/Results/..'


    # log_avg_score_r1 = np.load(folder_rl1 + '/Saved_models/log_average_score.npy')
    # log_avg_loses_r1 = np.load(folder_rl1 + '/Saved_models/log_avg_losses.npy')

    # log_avg_score_r2 = []
    # log_avg_loses_r2 = []

    # #log_avg_score_r2 = np.load(folder_rl2 + '/Saved_models/log_average_score.npy')
    # #log_avg_loses_r2 = np.load(folder_rl2 + '/Saved_models/log_avg_losses.npy')

    # log_avg_score = np.array(list(log_avg_score_r1) + list(log_avg_score_r2))
    # log_avg_loses = np.array(list(log_avg_loses_r1) + list(log_avg_loses_r2))

    #----- Random agent -----

    list_of_importance_metric_ran = np.load(folder_random + '/list_of_importance_metric.npy')
    list_of_patrol_metric_ran     = np.load(folder_random + '/list_of_patrol_metric.npy')
    list_of_reward_metric_ran     = np.load(folder_random + '/list_of_reward_metric.npy')
    list_of_perc_visited_ran      = np.load(folder_random + '/list_of_perc_visited.npy')

    #----- Greedy agent -----

    list_of_importance_metric_gr = np.load(folder_greedy + '/list_of_importance_metric.npy')
    list_of_patrol_metric_gr     = np.load(folder_greedy + '/list_of_patrol_metric.npy')
    list_of_reward_metric_gr     = np.load(folder_greedy + '/list_of_reward_metric.npy')
    list_of_perc_visited_gr      = np.load(folder_greedy + '/list_of_perc_visited.npy')

    #----- ZZ agent -----

    # list_of_importance_metric_zz = np.load(folder_zz + '/list_of_importance_metric.npy')
    # list_of_patrol_metric_zz     = np.load(folder_zz + '/list_of_patrol_metric.npy')
    # list_of_reward_metric_zz     = np.load(folder_zz + '/list_of_reward_metric.npy')
    # list_of_perc_visited_zz      = np.load(folder_zz + '/list_of_perc_visited.npy')

    #----- RL agent -----

    list_of_importance_metric_rl = np.load(folder_rl + '/list_of_importance_metric.npy')
    list_of_patrol_metric_rl     = np.load(folder_rl + '/list_of_patrol_metric.npy')
    list_of_reward_metric_rl     = np.load(folder_rl + '/list_of_reward_metric.npy')
    list_of_perc_visited_rl      = np.load(folder_rl + '/list_of_perc_visited.npy')

    #----- Plot Importance -----

    # fig1, ax1 = plt.subplots(dpi=180)
    # k = np.arange(len(list_of_importance_metric_ran))

    # ax1.set_yscale('log')
    # ax1.plot(k, list_of_importance_metric_ran, label='random')
    # ax1.plot(k, list_of_importance_metric_gr,  label='greedy')
    # ax1.plot(k, list_of_importance_metric_zz,  label='zz')
    # ax1.plot(k, list_of_importance_metric_rl,  label='rl')

    # ax1.grid('on')
    # ax1.legend()
    # ax1.set_xlabel(r'$k$')
    # ax1.set_ylabel(r'$Importance metric$')

    # fig1.savefig(folder + "/Importancemetric.jpg", dpi=180)
    # tikzplotlib.save(folder + "/Importancemetric.tex", figure=fig1)

    #----- Plot patrol metric  -----

    fig2, ax2 = plt.subplots(dpi=180)
    k = np.arange(len(list_of_patrol_metric_ran))

    #ax2.set_yscale('log')
    ax2.plot(k, list_of_patrol_metric_ran, label='random')
    ax2.plot(k, list_of_patrol_metric_gr,  label='greedy')
    #ax2.plot(k, list_of_patrol_metric_zz,  label='zz')
    ax2.plot(k, list_of_patrol_metric_rl,  label='rl')

    ax2.grid('on')
    ax2.legend()
    ax2.set_xlabel(r'$k$')
    ax2.set_ylabel(r'$Patrol metric$')

    # fig2.savefig(folder + "/Patrolmetric.jpg", dpi=180)
    # tikzplotlib.save(folder + "/Patrolmetric.tex", figure=fig2)


    #----- Plot reward metric -----

    fig3, ax3 = plt.subplots(dpi=180)
    k = np.arange(len(list_of_importance_metric_rl))

    #ax3.set_yscale('log')
    ax3.plot(k, np.array(list_of_reward_metric_ran)/10, label='random')
    ax3.plot(k, np.array(list_of_reward_metric_gr)/10,  label='greedy')
    #ax3.plot(k, np.array(list_of_reward_metric_zz)/10,  label='zz')
    ax3.plot(k, np.array(list_of_reward_metric_rl)/10,  label='rl')

    ax3.grid('on')
    ax3.legend()
    ax3.set_xlabel(r'$k$')
    ax3.set_ylabel(r'$Reward metric$')

    # fig3.savefig(folder + "/Rewardmetric.jpg", dpi=180)
    # tikzplotlib.save(folder + "/Rewardmetric.tex", figure=fig3)

    #----- Plot perentage metric -----

    fig4, ax4 = plt.subplots(dpi=180)
    k = np.arange(len(list_of_perc_visited_ran))

    #ax3.set_yscale('log')
    ax4.plot(k, list_of_perc_visited_ran, label='random')
    ax4.plot(k, list_of_perc_visited_gr,  label='greedy')
    #ax4.plot(k, list_of_perc_visited_zz,  label='zz')
    ax4.plot(k, list_of_perc_visited_rl,  label='rl')

    ax4.grid('on')
    ax4.legend()
    ax4.set_xlabel(r'$k$')
    ax4.set_ylabel(r'$Percentage$')

    # fig4.savefig(folder + "/perc.jpg", dpi=180)
    # tikzplotlib.save(folder + "/perc.tex", figure=fig4)

    #----- High percentage -----

    perc_array = np.array([0.0, 0.0, 0.0, 0.0])
    list_of_diff_rl = []
    list_of_diff_gr = []

    for i in range(len(list_of_patrol_metric_ran)):

        ran = list_of_patrol_metric_ran[i]
        gr  = list_of_patrol_metric_gr[i]
        #zz  = list_of_patrol_metric_zz[i]
        rl  = list_of_patrol_metric_rl[i]

        if ran >= gr and ran >= rl:

            perc_array[0] += 1
            max_elem = ran

        if gr >= ran and gr >= rl:

            perc_array[1] += 1
            max_elem = gr

        if rl >= ran and rl >= gr:

            perc_array[3] += 1
            max_elem = rl

        list_of_diff_rl.append(-rl+max_elem)
        list_of_diff_gr.append(-gr+max_elem)

    perc_array /= np.sum(perc_array)
    print("perc: ",perc_array)

    fig5, ax5 = plt.subplots(dpi=180)
    k = np.arange(len(list_of_diff_rl))

    ax5.plot(k, list_of_diff_rl, label='rl')
    ax5.plot(k, list_of_diff_gr, label='gr')

    ax5.grid('on')
    ax5.legend()
    ax5.set_xlabel(r'$k$')
    ax5.set_ylabel(r'$max-drone$')

    # fig4.savefig(folder + "/diff.jpg", dpi=180)
    # tikzplotlib.save(folder + "/diff.tex", figure=fig5)

    print("Integrals: ",np.sum(list_of_diff_rl), np.sum(list_of_diff_gr))
    print("Avg C: ",np.sum(list_of_patrol_metric_ran)/2000, np.sum(list_of_patrol_metric_gr)/2000, np.sum(list_of_patrol_metric_rl)/2000)
    print("Avg R: ",np.sum(list_of_reward_metric_ran)/20000, np.sum(list_of_reward_metric_gr)/20000, np.sum(list_of_reward_metric_rl)/20000)
```
