```python
import torch
import torch.nn as nn
import torch.optim as optim
import random

# Paramètres
vocab_size = 1000       # Vocabulaire fictif
embedding_dim = 50      # Dimension des embeddings
hidden_size = 64        # Taille cachée de la GRU
num_classes = 2         # Exemple : positif (1) ou négatif (0)
sequence_length = 10    # Longueur fixe des séquences
batch_size = 16
num_epochs = 100

# 1. Création de données factices
def generate_fake_data(batch_size):
    # Séquence de "mots" aléatoires encodés entre 0 et vocab_size
    inputs = torch.randint(0, vocab_size, (batch_size, sequence_length))
    labels = torch.randint(0, num_classes, (batch_size,))
    return inputs, labels

# 2. Modèle GRU
class GRUClassifier(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_size, num_classes):
        super(GRUClassifier, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.gru = nn.GRU(embedding_dim, hidden_size, batch_first=True)
        self.fc = nn.Linear(hidden_size, num_classes)

    def forward(self, x):
        embedded = self.embedding(x)  # (batch, seq_len, embed_dim)
        out, h_n = self.gru(embedded)  # h_n: (1, batch, hidden_size)
        last_hidden = h_n.squeeze(0)   # On prend le dernier état
        return self.fc(last_hidden)

# 3. Entraînement
model = GRUClassifier(vocab_size, embedding_dim, hidden_size, num_classes)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

for epoch in range(num_epochs):
    inputs, labels = generate_fake_data(batch_size)
    
    outputs = model(inputs)
    loss = criterion(outputs, labels)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}")
```

```python
def evaluate(model, num_batches=10):
    model.eval()
    correct = 0
    total = 0

    with torch.no_grad():
        for _ in range(num_batches):
            inputs, labels = generate_fake_data(batch_size)
            outputs = model(inputs)
            _, predicted = torch.max(outputs, 1)
            correct += (predicted == labels).sum().item()
            total += labels.size(0)
    
    accuracy = correct / total
    print(f"Accuracy on test set: {accuracy * 100:.2f}%")
    model.train()
```

```python
# Après la boucle d'entraînement :
evaluate(model, num_batches=20)
```
